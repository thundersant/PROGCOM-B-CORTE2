```python
import random


class TicTacToe:

    def __init__(self):
        self.tablero = []

    def crear_tablero(self):
        for i in range(3):
            fila = []
            for j in range(3):
                fila.append('-')
            self.tablero.append(fila)

    def selección_jugador(self):
        return random.randint(0, 1)

    def marcar_lugar(self, fila, column, jugador):
        self.tablero[fila][column] = jugador

    def jugador_ganador(self, jugador):
        gana = None

        n = len(self.tablero)

       #comprobar fila
        for i in range(n):
            gana = True
            for j in range(n):
                if self.tablero[i][j] != jugador:
                    gana = False
                    break
            if gana:
                return gana

        # comprobar columnas
        for i in range(n):
            gana = True
            for j in range(n):
                if self.tablero[j][i] != jugador:
                    gana = False
                    break
            if gana:
                return gana

        # diagonales
        gana = True
        for i in range(n):
            if self.tablero[i][i] != jugador:
                gana = False
                break
        if gana:
            return gana

        gana = True
        for i in range(n):
            if self.tablero[i][n - 1 - i] != jugador:
                gana = False
                break
        if gana:
            return gana
        return False

        for fila in self.tablero:
            for item in fila:
                if item == '-':
                    return False
        return True

    def tablero_lleno(self):
        for fila in self.tablero:
            for item in fila:
                if item == '-':
                    return False
        return True

    def cambio_turno_jugador(self, jugador):
        return 'X' if jugador == 'O' else 'O'

    def mostrar_tablero(self):
        for fila in self.tablero:
            for item in fila:
                print(item, end=" ")
            print()

    def inicio(self):
        self.crear_tablero()

        Jugador = 'X' if self.selección_jugador() == 1 else 'O'
        while True:
            print(f"Turno de {Jugador} Jugador")

            self.mostrar_tablero()

            # pidiendo el tamaño del tablero
            fila, column = list(
                map(int, input("Ingrese el numero de filas y columnas: ").split()))
            print()

            # cuadrando tablero
            self.marcar_lugar(fila - 1, column - 1, Jugador)

            # jugador gana o no
            if self.jugador_ganador(Jugador):
                print(f"el jugador {Jugador} gana el juego!")
                break

            # juego empatado o no
            if self.tablero_lleno():
                print("Juego empatado!")
                break

            # siguiente turno
            Jugador = self.cambio_turno_jugador(Jugador)

        # tablero final
        print()
        self.mostrar_tablero()


# iniciar juego
print('hola, para elegir la ubicacion debes usar un duo de numero separados por un espacio así:')
print('1 1, 1 2, 1 3')
print('2 1, 2 2, 2 3')
print('3 1, 3 2, 3 3')
print('') 
      
tic_tac_toe = TicTacToe()
tic_tac_toe.inicio()
```
