```python
import random


casillas = [ ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ]
HUMANO = 'X'
COMPUTADOR = '0'
primer_jugador = HUMANO
turno = 1
combinacion_ganar = [  [0, 1, 2], [3, 4, 5], [6, 7, 8], [0, 3, 6],
                    [1, 4, 7], [2, 5, 8], [0, 4, 8], [2, 4, 6], ]


def tablero(inicial=False):
    """ Imprime el tablero de juego. Si este es el comienzo del juego,
        imprime del 1 al 9 en los recuadros para mostrar a los jugadores cómo elegir un
        caja. De lo contrario, actualice cada casilla con X o 0 de las casillas [].
    """
    print(('''
{} | {} | {}
-----------
{} | {} | {}
-----------
{} | {} | {}

        ''').format(*([x for x in range(1, 10)] if inicial else casillas)))


def asignar_turno(jugador, turno):
  
    """ Crea un ciclo que sigue preguntando al jugador actual por
        su entrada hasta que se haga una elección válida.
    """

    while True:
        if jugador is COMPUTADOR:
            casilla = movida_computador()
        else:
            casilla = input('Jugador %s, escriba un número del 1 al 9 para seleccionar un cuadro: '  %  jugador )

            try:
                casilla = int(casilla) - 1 # resta 1 para sincronizar con casillas[] números de índice
            except ValueError:
                # no es entero
                print('Ese \' s no es un número válido, inténtalo de nuevo. \n ' )
                continue

        if casilla < 0 or casilla > 8:
            print('Ese número está fuera de rango, inténtalo de nuevo. \n ')
            continue

        if casillas[casilla] == ' ': # valor inicial
            casillas[casilla] = jugador # jugador actual asignado
            break
        else:
            print('Esa casilla ya está marcada, inténtalo de nuevo. \n ')


def movida_computador():
    """ Devuelve un número entero aleatorio de 0 a 8 """
    return random.randint(0,8)
        

def cambio_jugador(turno):
    """ Cambia de jugador en función de cuántos movimientos se hayan realizado.
        X comienza el juego, así que si este turno  es par, es el turno de 0.
    """
    jugador_actual = COMPUTADOR if turno % 2 == 0 else HUMANO
    return jugador_actual


def quien_gana(jugador, turno):
    """ Comprueba si hay una victoria (o un empate). Para cada combinación en combinacion_ganar[],
        cuenta cuantos de sus cuadrados correspondientes tienen la marca del jugador.
        Si el recuento de puntos de un jugador llega a 3, devuelve una victoria.
        Si no lo hace, y este ya es el turno #9, devuelve un empate.
        Si tampoco, devuelve Falso para que el juego continúe.
    """
    if turno > 4: # necesita al menos 5 movimientos antes de que sea posible ganar
        for combo in combinacion_ganar:
            puntaje = 0
            for index in combo:
                if casillas[index] == jugador:
                    puntaje += 1
                if puntaje == 3:
                    return 'ganó'

        if turno == 9:
            return 'empate'


def jugar(jugador, turno):
   """ Crea un ciclo que mantiene el juego en juego
        hasta que termine en victoria o empate
   """
   while True:
        asignar_turno(jugador, turno)
        tablero()
        resultado = quien_gana(jugador, turno)
        if resultado == 'ganó':
            print('Juego terminado. ¡%s gana! \n '  %  jugador)
            break
        elif resultado == 'empate':
            print('Juego terminado. Es un empate. \ n ')
            break
        turno += 1
        jugador = cambio_jugador(turno)


# a jugar: 
print ( ' \n \n ¡Bienvenido a Tic Tac Toe para dos jugadores!' )
tablero(inicial=True)
jugar(primer_jugador, turno)
```
